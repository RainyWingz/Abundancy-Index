```python
target = 10            # target abundancy index
search = 10^30          # length of search

# abundancy index of a number
def abun(n):
    return sigma(n, 1)/n

target = abun(target)

# get the first element of the list
def first_member(a):
    return a[0]

# initialize tree, with 1
# a is the value that t must be divisible by
# b is the value that t/a must be coprime to

abtree = [[1,1]]

def iter_tree():
    if len(abtree)==0:
        return 2         # 2 means tree closed

    # iterate tree
    a = abtree[0][0]
    b = abtree[0][1]
    abtree.pop(0)

    # special cases
    abund = abun(a)
    if abund == target:
        return 0         # 0 means found
    if abund > target:
        return 1         # 1 means too big


    # find possible factor
    pos_f = 1
    if a != 1:
        for p_factor in factor(a):
            if gcd(p_factor[0], b) == 1:
                pos_f = p_factor[0]
                break

    # multiply by factor
    if pos_f != 1:
        abtree.append([a*pos_f, b])
        abtree.append([a, b*pos_f])
        abtree.sort(key = first_member)
    # find new index
    else:
        new_index = denominator(target/abund)
        # multiply by 2, following conjecture of no odd perfect numbers
        if new_index == 1:
            new_index = 2
        # cannot share a factor with b
        if gcd(new_index, b) != 1:
            return 3            # 3 means bad index
        abtree.append([a*new_index, b])
        abtree.sort(key = first_member)

# run until search finished
while true:
    if len(abtree) == 0:
        print("Finished. Tree size: 0")
        break
    testing = abtree[0][0]
    if testing > search:
        break
    result = iter_tree()
    if result == 0:
        print(str(testing) + " satisfies this index!")

# finish
if len(abtree) != 0:
    print("Finished. Tree size: " + str(len(abtree)))
```
